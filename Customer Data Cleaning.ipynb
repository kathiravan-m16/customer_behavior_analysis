```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns #snake-casing
```

```python
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
df.columns
```

```python
#create age-group column
age_labels = ['Young_Adult', 'Adult', 'Middle_Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = age_labels)

df[['age', 'age_group']].head()
```

```python
df['frequency_of_purchases'].unique().tolist()
```

```python
#create purchasing_period column
period_labels = { 'Fortnightly': 14, 'Weekly': 7, 'Annually': 365, 'Quarterly': 90, 'Bi-Weekly': 14, 'Monthly': 30, 'Every 3 Months': 90}

df['purchasing_period'] = df['frequency_of_purchases'].map(period_labels)

df[['frequency_of_purchases', 'purchasing_period']].head() 
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop(['promo_code_used'], axis = 1)
df.columns
```
